```python
import pandas as pd
df = pd.read_csv(r"C:\Users\DS\csv\SalaryData_Train(1).csv")
df.head()
```

```python
svclassifier = SVC(kernel = 'rbf')
svclassifier.fit(X_train , y_train)
```
